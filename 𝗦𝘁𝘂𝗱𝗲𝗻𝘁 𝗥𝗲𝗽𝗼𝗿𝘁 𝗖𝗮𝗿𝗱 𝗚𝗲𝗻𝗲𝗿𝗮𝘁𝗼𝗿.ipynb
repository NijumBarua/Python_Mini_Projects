# ğğ®ğ­ğ­ğ¢ğ§ğ  ğšğ¥ğ¥ ğ’ğ­ğ®ğğğ§ğ­ğ¬ ğƒğšğ­ğš ğ“ğ¨ğ ğğ­ğ¡ğğ« (ğ‹ğ¢ğ¬ğ­ ğ¨ğŸ ğƒğ¢ğœğ­ğ¢ğ¨ğ§ğšğ«ğ¢ğğ¬)

```python
students = [
    {"name": "Nijum", "math": 85, "eng": 78, "science": 92},
    {"name": "Durjoy", "math": 90, "eng": 80, "science": 80},
    {"name": "Anik",  "math": 60, "eng": 40, "science": 75}
]
print("Total students:", len(students))
print(students)
```

# ğ‹ğğ­'ğ¬ ğŸğ¢ğ ğ®ğ«ğ ğ¨ğ®ğ­ ğ“ğ¨ğ­ğšğ¥ ğŒğšğ«ğ¤ğ¬ ğ¨ğŸ ğğšğœğ¡ ğ¬ğ­ğ®ğğğ§ğ­ğ¬

```python
for s in students:
    s["total"] = s["math"] + s["eng"] + s["science"]

# check if total is added
for s in students:
    print(s["name"], "â†’ total:", s["total"])
```

# ğğ¨ğ° ğ‹ğğ­ğ¬ ğ…ğ¢ğ§ğ ğğ®ğ­ ğ“ğ¡ğ ğ“ğ¨ğ©ğ©ğğ« (ğ‡ğ¢ğ ğ¡ğğ¬ğ­ ğ­ğ¨ğ­ğšğ¥)

```python
highest = -1     # à¦¶à§à¦°à§à¦¤à§‡ à¦–à§à¦¬ à¦›à§‹à¦Ÿ à¦®à¦¾à¦¨
topper_name = ""
for s in students:
    if s["total"] > highest:
        highest = s["total"]
        topper_name = s["name"]
print("Topper:", topper_name, "with", highest, "marks")
```

# ğ‹ğğ­'ğ¬ ğğ«ğ¢ğ§ğ­ ğğ®ğ­ ğ­ğ¡ğ ğ«ğğšğ¥ ğ‘ğğ©ğ¨ğ«ğ­ ğ‚ğšğ«ğ â™¦ï¸

```python
for s in students:
    print("----- Report Card -----")
    print("Name   :", s["name"])
    print("Math   :", s["math"])
    print("English:", s["eng"])
    print("Science:", s["science"])
    print("Total  :", s["total"])
    print("-----------------------")
print("\nOverall Topper â†’", topper_name, "with", highest)
```
